```python
import pandas as pd
from category_encoders import TargetEncoder
```

```python
train_df = pd.read_csv("../data/processed/processed_train_data.csv")
test_df = pd.read_csv("../data/processed/processed_test_data.csv")
new_df = pd.read_csv("../data/raw/new_data.csv")

pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.max_rows', None) # Show all rows
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
print("Train Data: " + train_df.date.min() + " === " + train_df.date.max())
print("Test Data: " + test_df.date.min() + " === " + test_df.date.max())
print("New Data: " + new_df.date.min() + " === " + new_df.date.max())
```

```python
train_df['date'] = pd.to_datetime(train_df['date'])
test_df['date'] = pd.to_datetime(test_df['date'])
new_df['date'] = pd.to_datetime(new_df['date'])
```

```python
train_df.info()
```

```python
def date_feature(df):
    df['year'] = df.date.dt.year
    df.insert(1, "year", df.pop("year"))
    df['quarter'] = df.date.dt.quarter
    df.insert(2, "quarter", df.pop("quarter"))
    df['month'] = df.date.dt.month
    df.insert(3, "month", df.pop("month"))

    return df

train_df = date_feature(train_df)
test_df = date_feature(test_df)
new_df = date_feature(new_df)
```

```python
train_df.head()
```

```python
zip_freq = train_df["zipcode"].value_counts()
```

```python
train_df["zipcode"].nunique()
```

```python
train_df["zipcode_freq"] = train_df["zipcode"].map(zip_freq)
test_df["zipcode_freq"] = test_df["zipcode"].map(zip_freq).fillna(0)
new_df["zipcode_freq"] = new_df["zipcode"].map(zip_freq).fillna(0)
```

```python
train_df.head()
```

```python
te = TargetEncoder(cols=["city_full"])

train_df["city_encoded"] = te.fit_transform(train_df["city_full"], train_df["price"])
test_df["city_encoded"] = te.transform(test_df["city_full"])
new_df["city_encoded"] = te.transform(new_df["city_full"])
```

```python
train_df[["city_full", "city_encoded"]].head()
```

```python
drop_cols = ["date", "city_full", "city", "zipcode", "median_sale_price"]
train_df.drop(columns=drop_cols, inplace=True)
test_df.drop(columns=drop_cols, inplace=True)
new_df.drop(columns=drop_cols, inplace=True)
```

```python
# Save the cleaned dataset
train_df.to_csv("../data/features/featured_train_data.csv", index=False)
test_df.to_csv("../data/features/featured_test_data.csv", index=False)
new_df.to_csv("../data/features/featured_new_data.csv", index=False)
print("Dataset saved successfully!")
```
