```python
import pandas as pd
pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.max_rows', None) # Show all rows
```

```python
df = pd.read_csv("/content/HouseTS.csv") # C:\Users\HP\Desktop\3mtt\Build-Deploy-Regression-Time-Series-ML-FastAPI-MLFlow-Docker-CI-CD-AWS\data\raw\HouseTS.csv
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df.date.dt.year.value_counts()
```

```python
new_df = df.sort_values("date")
new_df.head()
```

```python
train_df = df[new_df['date'] < "2020-01-01"]
test_df = new_df[(new_df['date'] >= "2020-01-01") & (new_df['date'] < "2021-12-31")]
eval_df = df[df['date'] > "2021-12-31"]
```

```python
print(train_df.shape)
print(test_df.shape)
print(eval_df.shape)
```

```python
train_df.to_csv("train_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)
eval_df.to_csv("new_data.csv", index=False)
```
